```python
import matplotlib.pyplot as plt
import numpy as np

# We fix the corruption mode to be honest majority
# We fix the number of parameters to be 100k
# Data: Each category has an array of values representing different groups
categories = ['10', '20', '30', '40']

# Data for each category represented as an array for each bar
data = {
    '10': table_ho_time.loc[0],
    '20': table_ho_time.loc[1],
    '30': table_ho_time.loc[2],
    '40': table_ho_time.loc[3],
}

# Reverse the data order (to match the bar chart order)
categories.reverse()

# Reverse the data arrays for each category
# for category in data:
#     data[category].reverse()

# Create figure and axes
fig, ax = plt.subplots(figsize=(10, 4))

# Narrow the bars by adjusting the width
bar_width = 0.5  # The width of each bar

# Define the colors for each group
colors = ['skyblue', 'lightgreen', 'salmon', 'purple', 'red']

# Prepare a list for the 'left' positions to stack bars
left_position = np.zeros(len(categories))

# Loop over each group and create the stacked bar for each category
for i, (group_name, group_values) in enumerate(zip(['InCom', 'B2A', 'L2', 'Agg', 'MAC Check'], zip(*data.values()))):
    ax.barh(categories, group_values, left=left_position, height=bar_width, color=colors[i], label=group_name)
    left_position += np.array(group_values)  # Update left position for the next stack

# Add labels and title
ax.set_ylabel('#Clients')
ax.set_xlabel('Runtime (sec)')
ax.set_title('End-to-end runtime')

# Add gridlines for better readability
ax.grid(axis='x', linestyle='--', alpha=0.7)

# Place the legend at the bottom and outside of the chart
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=5)

# Adjust layout for better spacing
plt.tight_layout()

# Save the plot as a PDF
plt.savefig("stacked_bar_chart_with_groups_as_arrays.pdf", format='pdf')

# Display the plot
# plt.show()
```
